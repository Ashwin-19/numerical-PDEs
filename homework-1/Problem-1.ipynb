$\text{Problem 1: Modeling and analytical solution}$

$\text{(a) }\frac{dQ}{dt} = 0.01 \cdot 300 - \frac{Q \cdot 300}{10^{6}} $
<br/>
$\implies \frac{dQ}{dt} = 3 - 3Q \cdot 10^{-4}$
$\text{Since the pond is initially free of an undesirable chemical, }Q(0) = Q_{0} = 0$

$\text{(b) }\frac{dQ}{dt} = 3 - 3Q \cdot 10^{-4}$
<br/>
$\implies \frac{dQ}{3 - 3Q \cdot 10^{-4}} = dt$
<br/>
$\implies \int_{Q_{0}}^{Q} \frac{dQ}{3 - 3Q \cdot 10^{-4}} = \int_{0}^{t} dt$
<br/>
$\implies  \frac{1}{3} [\frac{\log(1-Q \cdot 10^{-4})}{-10^{-4}}]_{Q_{0}}^{Q} = [t]_{0}^{t}$
<br/>
$\implies  \frac{1}{3} [\frac{\log(1-Q \cdot 10^{-4})}{-10^{-4}}]_{0}^{Q} = [t]_{0}^{t}   \because Q_{0} = 0$
<br/>
$\implies  \frac{1}{-3 \cdot 10^{-4}} \log(\frac{1-Q \cdot 10^{-4}}{1-0 \cdot 10^{-4}}) = t$
<br/>
$\implies \log(1-Q \cdot 10^{-4}) = -3 \cdot 10^{-4}t$
<br/>
$\implies 1-Q \cdot 10^{-4} = e^{-3 \cdot 10^{-4}t}$
<br/>
$\implies Q = 10^{4}(1-e^{-3 \cdot 10^{-4}t})$
<br/>
$\implies Q(t) = 10^{4}(1-e^{-3 \cdot 10^{-4}t})$

$\therefore Q\text{(t = 1 year)} = 10^{4}(1-e^{-3 \cdot 10^{-4} \cdot 365 \cdot 24})$
<br/>
$\implies Q\text{(t = 1 year)} = 10^{4}(1-e^{-2.6280}) = 9.278 \cdot 10^{3} g$
<br/>
$\implies Q\text{(t = 1 year)} = 9.278\text{ kg of chemical}$


$\text{(c) } \frac{dQ}{dt} = - \frac{Q \cdot 300}{10^{6}}$
$\implies \frac{dQ}{dt} = -3 \cdot 10^{-4} Q$
$\text{Since the pond has 9.278 kg of an undesirable chemical after one year, }$
$Q(0) = Q_{0} = 9.278 \cdot 10^{3} g$

$\text{(d) } \frac{dQ}{dt} = -3 \cdot 10^{-4} Q$
<br/>
$\implies \frac{dQ}{Q} = -3 \cdot 10^{-4} dt$
<br/>
$\implies \int_{Q_{0}}^{Q} \frac{dQ}{Q} = \int_{0}^{t} -3 \cdot 10^{-4} dt$
<br/>
$\implies [\log(Q)]_{Q_{0}}^{Q} = -3 \cdot 10^{-4} [t]_{0}^{t}$
<br/>
$\implies \log(\frac{Q}{Q_{0}}) = -3 \cdot 10^{-4}t$
<br/>
$\implies Q = Q_{0}e^{-3 \cdot 10^{-4}t}$
<br/>
$\implies Q(t) = Q_{0}e^{-3 \cdot 10^{-4}t}$
<br/>

$\therefore Q\text{(t = 1 year)} = Q_{0}e^{-3 \cdot 10^{-4} \cdot 365 \cdot 24}$
<br/>
$\implies Q\text{(t = 1 year)} = 9.278 \cdot 10^{3} \cdot e^{-3 \cdot 10^{-4} \cdot 365 \cdot 24}$
<br/>
$\implies Q\text{(t = 1 year)} = 670.08 g \text{ of chemical.}$

$\text{(e) } Q(t) = Q_{0}e^{-3 \cdot 10^{-4}t}$
<br/>
$\implies \log(\frac{Q}{Q_{0}}) = -3 \cdot 10^{-4}t$
<br/>
$\implies t = \frac{1}{-3 \cdot 10^{-4}} \log(\frac{Q}{Q_{0}})$
<br/>
$\implies t = \frac{1}{-3 \cdot 10^{-4}} \log(\frac{10}{9278})$
<br/>
$\implies t = \frac{1}{3 \cdot 10^{-4}} \log(\frac{9278}{10})$
<br/>
$\implies t = \frac{1}{3 \cdot 10^{-4}} \cdot 6.832 = 2.277 \cdot 10^{4} hours$
<br/>
$\implies t = 2.6 \text{ years.}$
<br/>
$\therefore \text{it will take 2.6 years for Q(t) to be reduced to 10g}$

$\text{(f) We can write down our analytical solution as follows:}$
<br/>
$Q(t) = \begin{cases} 
      10^{4}(1-e^{-3 \cdot 10^{-4}t}) & t \leq 1 \text{ year} \\
      9278 \cdot e^{-3 \cdot 10^{-4}t} & t > 1 \text{ year} \\
   \end{cases}$
$\text{We compute Q(t) from t = 0 till t = 3 years, and then plot Q(t) vs t.}$

```python
'''Plot for Q(t) vs t'''
from __future__ import division
import numpy as np
import scipy as sp
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import numpy.linalg as npla
import scipy.linalg as spla

def f1(t):
    ans = 1 - np.exp(-0.0003*t)
    return 10000*ans

def f2(t):
    ans = np.exp(-0.0003*t)
    return 9277.772365836792*ans

total_hours = 3*365*24
step = 365*24
hours = 0

x = np.linspace(1,total_hours,total_hours)
y = []

while hours <= step:
    y.append(f1(hours))
    hours += 1

total_hours -= step
hours = 1

while hours < total_hours:
    y.append(f2(hours))
    hours += 1

years = [365*24*i for i in range(1,4,1)]

figure(num=None, figsize=(8, 6), dpi=80, facecolor='w', edgecolor='k')
plt.title("Q(t) vs t")
plt.grid(color='grey', linestyle='dashed', linewidth=0.5)
plt.plot(x,y,label="Q(t)")
plt.axvline(years[0], label = "t = 1 year",color='yellow',linestyle='dashed')
plt.axvline(years[1], label = "t = 2 years",color='orange',linestyle='dashed')
plt.axvline(years[2], label = "t = 3 years",color='red',linestyle='dashed')
plt.legend(loc='best')
plt.ylabel("Q(t) (in grams)")
plt.xlabel("t (in hours)")
plt.show()
```
